Importing the Libraries and Data



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Spotify_Youtube Dataset.csv")
```

Checking the dataset




```python
data.info()
```

Dropping unwanted column

```python
data.drop(columns=['Unnamed: 0','Url_youtube','Uri','Url_spotify'], inplace = True)
data
```

Checking Null values

```python
data.isna().sum()
```

Replacing and replacing null values from dataframe

For likes and comment replacing null with 0 assuming no likes and comments

For other records we will simply drop the records


```python
data['Likes'] = data['Likes'].fillna(0)
data['Comments'] = data['Comments'].fillna(0)
data.dropna(inplace = True)
data.isna().sum()
```

Q.1 Top 10 Artists - with the Highest Views on YouTube?


```python
grouped_artist = data.groupby('Artist')['Views'].sum()
artist_Sorted = grouped_artist.sort_values(ascending=False)
artist_Sorted.head(10)
```

Q.2 Top 10 Tracks - with the Highest Streams on Spotify?



```python
track_with_streams = data.sort_values(by = 'Stream' , ascending =False)
track_with_streams[['Track','Stream']].head(10)
```

Q.3 What are the most common Album Types on Spotify? How many tracks belong to each album type?


```python
data.Album_type.unique()
al_type = data['Album_type'].value_counts()
```

```python
plt.pie(al_type , labels = al_type.index , autopct = "%1.1f%%" , startangle = 60 , shadow=True , explode =(0.05,0.05,0.05))
plt.show()
```

Q.4 How do the Average Views, Likes, and Comments are compared between different Album Types?


```python
grouped_ablumtype = data.groupby('Album_type')[['Likes','Views','Comments']].mean()
```

Q.5 Top 5 YouTube Channels -  based on the Views?

```python
grouped_channel = data.groupby('Channel')['Views'].sum()
sorted_channel = grouped_channel.sort_values(ascending = False).head(5)
sorted_channel
```

```python
sorted_channel = sorted_channel.reset_index()
sorted_channel.head(10)
type(sorted_channel)
```

```python
sns.barplot(x='Views',y='Channel' , data=sorted_channel)
plt.title("Top 5 Channel Based on Views")
plt.show()
```

Q.6 The Top Most Track -  based on Views?


```python
sorted_toptrack = data.sort_values(by='Views' , ascending=False)
sorted_toptrack[['Track' ,'Views']].head(1)
```

Q.7 Which Top 7 Tracks have the highest Like-to-View ratio on YouTube?

```python
track_lvr = data[['Track','Likes','Views']]
track_lvr['lv_ratio'] = data['Likes']/data['Views']*100
```

```python
track_lvr.head()
```

```python
track_lvr.sort_values(by ='lv_ratio' , ascending = False).head(7)
```

Q.8 Top Albums having the Tracks with Maximum Danceability ?

```python
grouped_albums = data.groupby('Album')['Danceability'].sum().sort_values(ascending = False)
grouped_albums.head()
```
